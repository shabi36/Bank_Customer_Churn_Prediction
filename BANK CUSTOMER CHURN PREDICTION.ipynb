```python
x = dataset.drop( columns= ["Exited"])
y =  (dataset["Exited"])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split( x , y , test_size= 0.2 , random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder , StandardScaler
from sklearn.metrics import  accuracy_score , confusion_matrix , precision_score
```

```python
from sklearn.cluster import KMeans , DBSCAN , HDBSCAN
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier,AdaBoostClassifier,ExtraTreesClassifier
from sklearn.svm import SVC
```

```python
dataset.info()
```

```python
ct = ColumnTransformer ( transformers= [("Col trans" , OneHotEncoder ( sparse_output= False , drop = "first"),
                                          [1,2,12])] , remainder= "passthrough"
                          )
```

```python

knn= KNeighborsClassifier (n_neighbors=8)

pipe = Pipeline ([
    ("step1" , ct),
     ("step2" , StandardScaler()),
    ("step3" , knn)
])


pipe.fit (x_train , y_train)

y_predict = pipe.predict (x_test)



print("\nTRAIN ",pipe.score(x_train,y_train))
print("TEST", pipe.score(x_test,y_test))

print("\nCONFUSION MATRIX\n", confusion_matrix(y_test , y_predict))

print("\nACCURACY SCORE", accuracy_score(y_test , y_predict))
print("PRECISION SCORE", precision_score(y_test , y_predict))
```

```python
dt= DecisionTreeClassifier (max_depth= 4)

pipe = Pipeline ([
    ("step1" , ct),
     ("step2" , StandardScaler()),
    ("step3" , dt)
])


pipe.fit (x_train , y_train)

y_predict = pipe.predict (x_test)



print("\nTRAIN ",pipe.score(x_train,y_train))
print("TEST", pipe.score(x_test,y_test))

print("\nCONFUSION MATRIX\n", confusion_matrix(y_test , y_predict))

print("\nACCURACY SCORE", accuracy_score(y_test , y_predict))
print("PRECISION SCORE", precision_score(y_test , y_predict))
```

```python
from sklearn.ensemble import VotingClassifier

knn= KNeighborsClassifier (n_neighbors=8)

dt= DecisionTreeClassifier (max_depth= 4)

rf= RandomForestClassifier (n_estimators=300,
                              random_state=3,
                              max_samples=0.5,
                              max_features=0.75,
                              max_depth=25)

et=  ExtraTreesClassifier (n_estimators=500,
                              random_state=3,
                              max_samples=0.5,
                              max_features=0.75,
                              max_depth=15,
                             bootstrap=True)

ad=  AdaBoostClassifier (n_estimators=50,
                              learning_rate= 1)

gd=  GradientBoostingClassifier (n_estimators=200)



vc = VotingClassifier ([ ("rf" , rf ), ("gd" , gd) , ("et" , et) , ("dt" , dt) , ("knn" , knn) , ( "ad", ad)] , weights = [5,5,5,5,5,5])



```

```python
pipe = Pipeline ([
    ("step1" , ct),
      ("step2" , StandardScaler()),
    ("step3" , vc)
])
```

```python
pipe.fit (x_train , y_train)
```

```python
y_predict = pipe.predict (x_test)
```

```python


print("\nTRAIN ",pipe.score(x_train,y_train))
print("TEST", pipe.score(x_test,y_test))

print("\nCONFUSION MATRIX\n", confusion_matrix(y_test , y_predict))

print("\nACCURACY SCORE", accuracy_score(y_test , y_predict))
print("PRECISION SCORE", precision_score(y_test , y_predict))
```
